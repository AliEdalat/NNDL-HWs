```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = [0.32, 0.33, 0.28, 0.47, 0.66, 0.80, 0.4, 0.33, 0.1, 0.26]
```

```python
def act(x):
  if x < 0:
    return 0
  elif x>=0 and x<2:
    return x
  else:
    return 2
```

```python
def get_w(i,j,r1,r2,w1,w2):
  if j <= i+r1 and j >= i-r1:
    return w1
  elif (j > i+r1 and j <= i+r2) or (j < i-r1 and j >= i-r2):
    return -1*w2
  else:
    return 0
```

```python
def mexican_hat(x,r1, r2, w1=0.6, w2=0.4):
  xt = x.copy()
  it = 0
  while it < 3:
    for i in range(len(xt)):
      res = 0
      for j in range(len(xt)):
        res += get_w(i,j,r1,r2,w1,w2)*xt[j]
      xt[i] = res
    xt = list(map(act, xt)) 
    print(xt)
    plt.plot(xt)
    plt.xlabel('xi')
    plt.ylabel('x')
    plt.grid()
    plt.show()
    it += 1
```

```python
r1, r2 = 0, float('inf')
mexican_hat(X,r1, r2, w1=0.5, w2=0.4)
```

```python
r1, r2 = 1, 5
mexican_hat(X,r1, r2, w1=0.5, w2=0.4)
```
